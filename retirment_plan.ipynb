```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use(
income = (17000+15000) * 12
expenses = 15000  * 12
saving_rate = (income - expenses)/income
interst_rate = 0.07
investment = 500000
annual_rate = investment * interst_rate 
year = 2021
withdrawal_rate = 0.1
```

```python
years = []
yearly_income = []
yearly_expense= []
yearly_invested_ammount = []
total_invested_ammount= []
annual_returns= []

# annual_return = int(investment * interst_rate)
# investment = income-expenses
#years.append(year)
#yearly_expense.append(expenses)
#yearly_income.append(income)
#yearly_invested_ammount.append(income-expenses)
#total_invested_ammount.append(investment)
#annual_returns.append(annual_return)
invested_years = 20
annual_return = 0
```

```python
for i in range(0,invested_years - 1):
    investment = int(income - expenses + annual_return + investment)
    annual_return = int(investment * interst_rate)
    years.append(year+i+1)
    yearly_income.append(income)
    yearly_expense.append(expenses)
    yearly_invested_ammount.append(income-expenses)
    annual_returns.append(annual_return)
    total_invested_ammount.append(investment)
```

```python
df = pd.DataFrame()
df['year'] = years
df['yearly_income'] = yearly_income
df['yearly_expense'] = yearly_expense
df['yearly_Gain'] = yearly_invested_ammount
df['total_invested_ammount'] = total_invested_ammount
df['annual_returns'] = annual_returns
yearly_withdraw_ammount = np.array(total_invested_ammount) * withdrawal_rate
df['yearly_withdraw_ammount'] = yearly_withdraw_ammount
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['year'],df['yearly_expense'] ,label = "yearly expenses")
plt.plot(df['year'],df['annual_returns'] ,label = "yearly return ammount")
plt.title("years untill retirment")
plt.xlabel("Years")
plt.xticks(df['year'],rotation =45)
plt.legend()
plt.show()
df
```
